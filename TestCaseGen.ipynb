```python
# =========================================
# 1. INSTALL DEPENDENCIES
# =========================================
!pip install -q openai

# =========================================
# 2. IMPORTS
# =========================================
import os
from getpass import getpass
from openai import OpenAI

# =========================================
# 3. API KEY
# =========================================
os.environ["OPENAI_API_KEY"] = getpass("Enter your OpenAI API Key: ")
client = OpenAI()

# =========================================
# 4. MEMORY SETUP
# =========================================

# Short-term memory (current session)
conversation_history = []

# Long-term memory (simulated)
long_term_memory = []

# Semantic memory (domain knowledge)
semantic_memory = """
Testing Guidelines:
- Always include positive, negative, and edge test cases
- Validate input fields (empty, invalid, boundary)
- Include error handling scenarios
- Consider security and performance basics
"""

# =========================================
# 5. AGENT FUNCTION (ReAct Style)
# =========================================
def generate_test_cases(user_input):
    global conversation_history, long_term_memory

    # Combine all memory
    memory_context = f"""
    Short-Term Memory:
    {conversation_history}

    Long-Term Memory:
    {long_term_memory}

    Semantic Memory:
    {semantic_memory}
    """

    # ReAct-style prompt
    prompt = f"""
    You are a QA Test Engineer Agent.

    Follow this process:
    1. Understand the requirement
    2. Think step-by-step (Reason)
    3. Generate structured test cases (Act)

    {memory_context}

    Requirement:
    {user_input}

    Output format:
    - Test Case ID
    - Description
    - Type (Positive/Negative/Edge)
    - Priority (High/Medium/Low)
    """

    try:
        response = client.chat.completions.create(
            model="gpt-4o-mini",
            messages=[{"role": "user", "content": prompt}],
            temperature=0
        )

        result = response.choices[0].message.content

        # Update memories
        conversation_history.append({"user": user_input, "agent": result})
        long_term_memory.append(user_input)

        return result

    except Exception as e:
        return f"Error: {e}"

# =========================================
# 6. RUN AGENT LOOP
# =========================================
print("🤖 AI Test Case Generator Agent (type 'exit' to stop)\n")

while True:
    user_input = input("Enter business requirement:\n")

    if user_input.lower() == "exit":
        break

    output = generate_test_cases(user_input)

    print("\n📄 Generated Test Cases:\n")
    print(output)
    print("\n" + "="*50 + "\n")
```
